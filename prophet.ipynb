# Setup and Train

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from pandas.plotting import register_matplotlib_converters

%matplotlib inline
%config InlineBackend.figure_format='retina'

register_matplotlib_converters()
sns.set(style='whitegrid', palette='muted', font_scale=1.5)

figsize = (22, 10)
plt.rcParams['figure.figsize'] = figsize
```

```python
from prophet import Prophet
```

```python
df = pd.read_csv('fb_close.csv', parse_dates=['Date'])
df.head()
```

```python
df.rename(columns={'Date': 'ds', 'Close': 'y'}, inplace=True)
```

```python
df.tail()
```

```python
model = Prophet()
```

```python
model.fit(df)
```

```python
# CARE! The data only consists of business days,
#  therefore, creating future dates of only business days
future = model.make_future_dataframe(periods=60, freq='B')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
forecast.columns
```

```python
forecast.shape
```

```python
fig = model.plot(forecast, figsize=figsize)
```

# Zoomed Figure

```python
forecast_zoomed = forecast.copy()
forecast_zoomed = forecast_zoomed[['ds', 'yhat']]
forecast_zoomed.tail()
```

```python
forecast_zoomed = pd.merge(forecast_zoomed, df, on='ds', how='outer', sort=True)
forecast_zoomed = forecast_zoomed.set_index('ds').iloc[-100:]
forecast_zoomed.tail()
```

```python
forecast_zoomed.plot(title='Forecast (Zoomed)')
plt.show()
```

# Evaluation

```python
fig2 = model.plot_components(forecast)
```

```python
from prophet.diagnostics import cross_validation
df_cv = cross_validation(model, initial='730 days', period='180 days', horizon = '365 days')
```

```python
df_cv.head()
```

```python
from prophet.diagnostics import performance_metrics
df_p = performance_metrics(df_cv)
df_p.head()
```

```python
from prophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(df_cv, metric='rmse')
```
